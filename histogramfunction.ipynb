```python
import cv2
```

```python
image=cv2.imread('coco.jpg',0)
```

```python
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.hist(image.ravel(),256,[0,256])
plt.show()
```

```python
equalize=cv2.equalizeHist(image)
```

```python
equalize
```

```python
plt.hist(equalize.ravel(),256,(0,256))
plt.show()
```

```python
cv2.imshow('equalized',equalize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
result=np.hstack((equalize,image))
```

```python
plt.title('Equalized ----------------------Before Equalization')
plt.imshow(result)
```
