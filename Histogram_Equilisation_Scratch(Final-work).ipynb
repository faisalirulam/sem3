```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('coco.jpg',0)
```

```python
f=cv2.resize(img,dsize=(100,100))
p=np.array(f)
print(p.shape)
plt.imshow(p)
```

```python
s=np.unique(f)
print(s)
len(s)
```

```python
r,c = f.shape
```

```python
count=0
d=0
arr=[]
for k in range(len(s)):
    for i in range(r):
        for  j in range(c):
            if s[k] == f[i][j]:
                count+=1 
                d+=1
                arr.append(count)
                if count>1:
                    u=len(arr)
                    arr[u-2]=count
                    del arr[u-1]
    count=0
print("Total number of pixels:",d)
print("Number of pixels:",arr)
print("Number of array Elements:",len(arr))
```

```python
em=[]
j=0
for i in range(0,len(arr)):
    j+=arr[i]
    j=j/d
    j=round(j*np.max(f))
    em.append(j)  
print(em)
```

```python
for i in range(r):
    for j in range(c):
        if f[i][j]==s[i]:
            f[i][j]=em[i]
```
